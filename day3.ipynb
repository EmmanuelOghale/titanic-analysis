```python
import pandas as pd
import numpy as np

df = pd.read_csv("titanic.csv")
print(df.head())
```

```python
print(df.shape)      # how many rows and columns?
print(df.columns)    # what columns exist?
print(df.dtypes)     # what data type is each column?
print(df.describe()) # statistics on all number columns
```

```python
print(df.isnull().sum())
```

```python
# Fix all three
df["Age"] = df["Age"].fillna(df["Age"].mean())
df = df.drop("Cabin", axis=1)
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Check missing values again
print(df.isnull().sum())
```

```python
temps = np.array([28, 35, 22, 40, 18, 31])
print(f"{np.mean(temps):.2f} °C")  # average 
print(np.max(temps), '°C')     # highest
print(np.min(temps), '°C')     # lowest
print(np.std(temps), '°C')     # standard deviation
```

```python
F= (9*temps/5)+32
for c, f in zip(temps, F):
    print(f"{c} °C → {f:.1f} °F")
```

```python
print(temps[temps > 30])
```
