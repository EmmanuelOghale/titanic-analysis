```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Load the Titanic dataset
df = pd.read_csv("titanic.csv")

# Clean it — same as yesterday
df["Age"] = df["Age"].fillna(df["Age"].mean())
df = df.drop("Cabin", axis=1)
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

print("Ready to visualise!")
```

```python
# Simple line chart
x = [1, 2, 3, 4, 5]
y = [10, 25, 15, 30, 20]

plt.figure(figsize=(8, 4))      # canvas size — width, height in inches
plt.plot(x, y)                   # draw the line
plt.title("My First Line Chart") # title at top
plt.xlabel("X Axis")             # label for x axis
plt.ylabel("Y Axis")             # label for y axis
plt.show()                       # display it
```

```python
# Count survivors vs non-survivors
survival_counts = df["Survived"].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(["Did Not Survive", "Survived"],   # labels on x axis
        survival_counts,                    # heights of bars
        color=["#E74C3C", "#2ECC71"])       # red and green
plt.title("Titanic Survival Count")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
# Survival by passenger class
class_survival = df.groupby("Pclass")["Survived"].mean() * 100

plt.figure(figsize=(6, 4))
plt.barh(["1st Class", "2nd Class", "3rd Class"],
          class_survival,
          color=["#3498DB", "#9B59B6", "#E67E22"])
plt.title("Survival Rate by Class (%)")
plt.xlabel("Survival Rate %")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["Age"],           # x axis — age
            df["Fare"],          # y axis — fare paid
            alpha=0.4,           # transparency — 0 invisible, 1 solid
            color="#C544AD")
plt.title("Age vs Fare Paid")
plt.xlabel("Age")
plt.ylabel("Fare (£)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y,
         color="blue",        # line colour
         linewidth=2,         # thickness of line
         linestyle="--",      # dashed line
         marker="o")          # dot at each data point
plt.title("Styled Chart", fontsize=16, fontweight="bold")
plt.xlabel("X Axis", fontsize=12)
plt.ylabel("Y Axis", fontsize=12)
plt.grid(True)                # adds gridlines
plt.tight_layout()            # prevents labels being cut off
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.bar(["Did Not Survive", "Survived"],
        survival_counts,
        color=["#E74C3C", "#2ECC71"])
plt.title("Titanic Survival Count")
plt.savefig("survival_chart.png",   # saves to your folder
            dpi=150,                  # quality — higher = sharper
            bbox_inches="tight")      # prevents cutting off
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df = pd.read_csv("titanic.csv")
df["Age"] = df["Age"].fillna(df["Age"].mean())
df = df.drop("Cabin", axis=1)
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

print("Ready!")
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Age"],
         bins=30,
         color="#3498DB",
         edgecolor="white")
plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.tight_layout()
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(6, 4))
sns.countplot(data=df,
              x="Survived",
              palette="Set2")
plt.title("Survival Count")
plt.xticks([0, 1], ["Did Not Survive", "Survived"])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df,
              x="Survived",
              hue="Sex",
              palette="Set1")
plt.title("Survival Count by Gender")
plt.xticks([0, 1], ["Did Not Survive", "Survived"])
plt.legend(title="Gender")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df,
            x="Pclass",
            y="Age",
            palette="pastel")
plt.title("Age Distribution by Passenger Class")
plt.xlabel("Passenger Class (1=First, 2=Second, 3=Third)")
plt.ylabel("Age")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df,
             x="Age",
             bins=30,
             kde=True,        # adds smooth curve on top
             color="#3498DB")
plt.title("Age Distribution with Distribution Curve")
plt.xlabel("Age")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df,
            x="Pclass",
            y="Survived",
            palette="Blues",
            hue="Pclass",
            errorbar=None)
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.tight_layout()
plt.show()
```

```python
# select_dtypes filters to number columns only
# because you can't correlate text columns like Name or Sex
correlation = df.select_dtypes(include="number").corr()
print(correlation)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation,
            annot=True,       # show the numbers inside each cell
            fmt=".2f",        # format numbers to 2 decimal places
            cmap="coolwarm",  # colour map — blue=negative, red=positive
            center=0)         # 0 is the neutral colour
plt.title("Correlation Heatmap — Titanic Dataset")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Generate correlation matrix
corr = df.select_dtypes(include="number").corr()

# Create a mask for the upper triangle — removes redundant cells
import numpy as np
mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(corr,
            mask=mask,          # show only lower triangle
            annot=True,
            fmt=".2f",
            cmap="coolwarm",
            center=0,
            square=True,        # make cells square shaped
            linewidths=0.5)     # add lines between cells
plt.title("Correlation Heatmap — Titanic Dataset")
plt.tight_layout()
plt.show()
```

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return f"₦{amount:,} deposited. New balance: ₦{self.balance:,}"

    def withdraw(self, amount):
        if amount > self.balance:
            return "Insufficient funds!"
        self.balance -= amount
        return f"₦{amount:,} withdrawn. New balance: ₦{self.balance:,}"

    def check_balance(self):
        return f"{self.owner}'s balance: ₦{self.balance:,}"

emmanuel_account = BankAccount("Emmanuel", 50000)
tunde_account = BankAccount("Tunde", 20000)

print(emmanuel_account.owner)
print(tunde_account.balance)
print(emmanuel_account.deposit(10000))
print(tunde_account.withdraw(25000))
print(emmanuel_account.check_balance())
print(type(emmanuel_account))
```

```python
class Student:
    def __init__(self, name, score):
        print(f"Creating student record for {name} with score {score}...")
        self.name = name
        self.score = score
        
    def get_grade(self):
        if self.score >= 90:
            return "A"
        elif self.score >= 80:
            return "B"
        elif self.score >= 50:
            return "C"
        else:
            return "F"
        
    def is_passing(self):
        return self.score >= 50
    
    def introduce(self):
        return f"Hi, I'm {self.name} and I scored {self.score}."
    

student1 = Student("Emmanuel", 85)
student2 = Student("Tunde", 42)
student3 = Student("Amaka", 91)

# Call all three methods on each object
for student in [student1, student2, student3]:
    print(student.introduce())
    print(f"Grade: {student.get_grade()}")
    print(f"Passing: {student.is_passing()}")
    print("---")
```

```python
data = {
    "Title":             ["Inception", "The Matrix", "Interstellar", "The Dark Knight", "Avengers"],
    "Year":              [2010, 1999, 2014, 2008, 2019],
    "Rating":            [8.8, 8.7, 8.6, 9.0, 8.4],
    "Genre":             ["Sci-Fi", "Sci-Fi", "Sci-Fi", "Action", "Action"],
    "Box_Office_Millions": [829, 463, 677, 1004, 2798]
}

df_movies = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.barh(df_movies["Title"], df_movies["Rating"], color=["#3498DB", "#E67E22", "#9B59B6", "#2ECC71", "#F39C12"])
plt.title("Movie Ratings")
plt.xlabel("Rating")
plt.ylabel("Movie Title")
plt.tight_layout()
plt.show()
```
